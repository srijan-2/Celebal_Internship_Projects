```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Titanic dataset
data_url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
titanic_df = pd.read_csv(data_url)
```

```python
# Display the first few rows of the dataset
print(titanic_df.head())

# Check the dimensions of the dataset
print('Shape:', titanic_df.shape)

# Check the data types of each column
print('Data Types:')
print(titanic_df.dtypes)

# Check for missing values
print('Missing Values:')
print(titanic_df.isnull().sum())
```

# Survival Analysis

```python
# Calculate the overall survival rate
survival_rate = titanic_df['Survived'].mean() * 100
print('Survival Rate: {:.2f}%'.format(survival_rate))

# Visualize the survival count
sns.countplot(x='Survived', data=titanic_df)
plt.title('Survival Count')
plt.show()
```

# Passenger Class Analysis:

```python
# Count the number of passengers in each class
passenger_count_by_class = titanic_df['Pclass'].value_counts().sort_index()
print('Passenger Count by Class:')
print(passenger_count_by_class)

# Visualize the passenger count by class
sns.countplot(x='Pclass', data=titanic_df)
plt.title('Passenger Count by Class')
plt.show()
```

# Gender Analysis:

```python
# Count the number of passengers by gender
passenger_count_by_gender = titanic_df['Sex'].value_counts()
print('Passenger Count by Gender:')
print(passenger_count_by_gender)

# Visualize the passenger count by gender
sns.countplot(x='Sex', data=titanic_df)
plt.title('Passenger Count by Gender')
plt.show()

# Survival rate by gender
survival_rate_by_gender = titanic_df.groupby('Sex')['Survived'].mean() * 100
print('Survival Rate by Gender:')
print(survival_rate_by_gender)

# Visualize the survival rate by gender
sns.barplot(x='Sex', y='Survived', data=titanic_df)
plt.title('Survival Rate by Gender')
plt.ylabel('Survival Rate')
plt.show()
```

# Age Analysis:

```python
# Summary statistics of age
age_summary = titanic_df['Age'].describe()
print('Summary Statistics of Age:')
print(age_summary)

# Distribution of age
sns.histplot(x='Age', data=titanic_df, kde=True)
plt.title('Distribution of Age')
plt.show()

# Age distribution by survival
plt.figure(figsize=(10, 5))
sns.histplot(x='Age', data=titanic_df, hue='Survived', kde=True)
plt.title('Age Distribution by Survival')
plt.legend(['Did not survive', 'Survived'])
plt.show()
```

# Family Size Analysis:

```python
# Calculate family size
titanic_df['FamilySize'] = titanic_df['SibSp'] + titanic_df['Parch'] + 1

# Count the number of passengers by family size
passenger_count_by_family_size = titanic_df['FamilySize'].value_counts().sort_index()
print('Passenger Count by Family Size:')
print(passenger_count_by_family_size)

# Visualize the passenger count by family size
plt.figure(figsize=(10, 5))
sns.countplot(x='FamilySize', data=titanic_df)
plt.title('Passenger Count by Family Size')
plt.show()

# Survival rate by family size
survival_rate_by_family_size = titanic_df.groupby('FamilySize')['Survived'].mean() * 100
print('Survival Rate by Family Size:')
print(survival_rate_by_family_size)

# Visualize the survival rate by family size
plt.figure(figsize=(10, 5))
sns.barplot(x='FamilySize', y='Survived', data=titanic_df)
plt.title('Survival Rate by Family Size')
plt.ylabel('Survival Rate')
plt.show()
```

# Fare Analysis:

```python
# Summary statistics of fare
fare_summary = titanic_df['Fare'].describe()
print('Summary Statistics of Fare:')
print(fare_summary)

# Distribution of fare
sns.histplot(x='Fare', data=titanic_df, kde=True)
plt.title('Distribution of Fare')
plt.show()

# Fare distribution by survival
plt.figure(figsize=(10, 5))
sns.histplot(x='Fare', data=titanic_df, hue='Survived', kde=True)
plt.title('Fare Distribution by Survival')
plt.legend(['Did not survive', 'Survived'])
plt.show()
```

# Embarked Port Analysis:

```python
# Count the number of passengers by embarked port
passenger_count_by_embarked = titanic_df['Embarked'].value_counts()
print('Passenger Count by Embarked Port:')
print(passenger_count_by_embarked)

# Visualize the passenger count by embarked port
sns.countplot(x='Embarked', data=titanic_df)
plt.title('Passenger Count by Embarked Port')
plt.show()

# Survival rate by embarked port
survival_rate_by_embarked = titanic_df.groupby('Embarked')['Survived'].mean() * 100
print('Survival Rate by Embarked Port:')
print(survival_rate_by_embarked)

# Visualize the survival rate by embarked port
sns.barplot(x='Embarked', y='Survived', data=titanic_df)
plt.title('Survival Rate by Embarked Port')
plt.ylabel('Survival Rate')
plt.show()
```

# Cabin Analysis:

```python
# Extract the cabin deck from the cabin number
titanic_df['CabinDeck'] = titanic_df['Cabin'].str[:1]

# Count the number of passengers by cabin deck
passenger_count_by_cabin_deck = titanic_df['CabinDeck'].value_counts()
print('Passenger Count by Cabin Deck:')
print(passenger_count_by_cabin_deck)

# Visualize the passenger count by cabin deck
sns.countplot(x='CabinDeck', data=titanic_df)
plt.title('Passenger Count by Cabin Deck')
plt.show()

# Survival rate by cabin deck
survival_rate_by_cabin_deck = titanic_df.groupby('CabinDeck')['Survived'].mean() * 100
print('Survival Rate by Cabin Deck:')
print(survival_rate_by_cabin_deck)

# Visualize the survival rate by cabin deck
sns.barplot(x='CabinDeck', y='Survived', data=titanic_df)
plt.title('Survival Rate by Cabin Deck')
plt.ylabel('Survival Rate')
plt.show()
```

# Name Analysis:

```python
# Extract the title from the name
titanic_df['Title'] = titanic_df['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)

# Count the number of passengers by title
passenger_count_by_title = titanic_df['Title'].value_counts()
print('Passenger Count by Title:')
print(passenger_count_by_title)

# Visualize the passenger count by title
plt.figure(figsize=(10, 5))
sns.countplot(x='Title', data=titanic_df)
plt.title('Passenger Count by Title')
plt.xticks(rotation=45)
plt.show()

# Survival rate by title
survival_rate_by_title = titanic_df.groupby('Title')['Survived'].mean() * 100
print('Survival Rate by Title:')
print(survival_rate_by_title)

# Visualize the survival rate by title
plt.figure(figsize=(10, 5))
sns.barplot(x='Title', y='Survived', data=titanic_df)
plt.title('Survival Rate by Title')
plt.ylabel('Survival Rate')
plt.xticks(rotation=45)
plt.show()
```

# Ticket Analysis:

```python
# Count the number of passengers by ticket frequency
passenger_count_by_ticket_frequency = titanic_df['Ticket'].value_counts()
print('Passenger Count by Ticket Frequency:')
print(passenger_count_by_ticket_frequency)

# Visualize the passenger count by ticket frequency
plt.figure(figsize=(10, 5))
sns.countplot(x=titanic_df['Ticket'].value_counts(), data=titanic_df)
plt.title('Passenger Count by Ticket Frequency')
plt.xticks(rotation=45)
plt.show()

# Survival rate by ticket frequency
survival_rate_by_ticket_frequency = titanic_df.groupby('Ticket')['Survived'].mean() * 100
print('Survival Rate by Ticket Frequency:')
print(survival_rate_by_ticket_frequency)

# Visualize the survival rate by ticket frequency
plt.figure(figsize=(10, 5))
sns.barplot(x=titanic_df['Ticket'].value_counts(), y='Survived', data=titanic_df)
plt.title('Survival Rate by Ticket Frequency')
plt.ylabel('Survival Rate')
plt.xticks(rotation=45)
plt.show()
```

# Age and Gender Combination:

```python
# Age and gender combination analysis
plt.figure(figsize=(10, 5))
sns.histplot(x='Age', data=titanic_df, hue='Sex', kde=True)
plt.title('Age Distribution by Gender')
plt.legend(['Male', 'Female'])
plt.show()

# Survival rate by age and gender combination
age_bins = [0, 18, 30, 50, np.inf]
age_labels = ['0-18', '18-30', '30-50', '50+']
titanic_df['AgeGroup'] = pd.cut(titanic_df['Age'], bins=age_bins, labels=age_labels)
survival_rate_by_age_gender = titanic_df.groupby(['AgeGroup', 'Sex'])['Survived'].mean() * 100
print('Survival Rate by Age and Gender Combination:')
print(survival_rate_by_age_gender)

# Visualize the survival rate by age and gender combination
plt.figure(figsize=(10, 5))
sns.barplot(x='AgeGroup', y='Survived', hue='Sex', data=titanic_df)
plt.title('Survival Rate by Age and Gender Combination')
plt.ylabel('Survival Rate')
plt.show()
```
